```python
fn= "/content/drive/MyDrive/I_have_a_dream (1).txt"
class text_static:
  def processLine(self,line,wordCounts):
   line=self.replacePunctuations(line)
   words=line.split()
   for word in words:
     if word in wordCounts:
       wordCounts[word]+=1
     else:
       wordCounts[word]=1

  def replacePunctuations(self,line):
    for ch in line:
      if ch in " ":
        line=line.replace(ch," ")

    return line

  def wordCounts(self,filename):
    f=open("/content/drive/MyDrive/I_have_a_dream (1).txt","r")
    lines=f.readlines()
    wordCounts={}
    for line in lines:
      self.processLine(line.lower(),wordCounts)
    return wordCounts

  def arrange_pairs(self,wordCounts):
    pairs=list(wordCounts.items())
    items=[[x,y] for(y,x) in pairs]
    items.sort()
    items={x:y for [y,x] in items}
    return items
```

```python
T=text_static()
```

```python
wordCounts=T.wordCounts(fn)
```
